```python
from simulation.simulator import Simulator
```

```python
scipy_sim = Simulator("scipy")
```

```python
def diffeq_logistic_growth(t, z, args):
    """
    Example diffeq for logistic growth.
    """
    a_x, b_x, a_y, b_y = args
    x, y = z

    dxdt = a_x * (b_x - x)
    dydt = a_y * (b_y - y)

    return [dxdt, dydt]
```

```python
a_x = 0.1
b_x = 100
a_y = 0.02
b_y = 300

y0_0 = 1
y1_0 = 2
```

```python
scipy_sim.generate_numeric_sol_ivp(
    diffeq_func=diffeq_logistic_growth,
    args=(a_x, b_x, a_y, b_y),
    ic=[y0_0, y1_0],
    ti=0,
    tf=500,
    dt=0.1,
)
```

```python
plots = scipy_sim.plot_numeric_sol_ivp()
```
